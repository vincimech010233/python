```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular el período del péndulo
def calcular_periodo(longitud, gravedad):
    return 2 * np.pi * np.sqrt(longitud / gravedad)

# Parámetros iniciales
longitudes = np.linspace(0.1, 10, 100)  # Longitudes de 0.1 a 10 metros
gravedad = 9.81  # Gravedad en m/s^2
error_medicion = 0.05  # 5% de error

# Cálcular los períodos sin error
periodos = [calcular_periodo(L, gravedad) for L in longitudes]

# Simular los errores
errores = error_medicion * np.random.randn(len(longitudes))
longitudes_erroneas = longitudes * (1 + errores)
periodos_erroneos = [calcular_periodo(L, gravedad) for L in longitudes_erroneas]

# Cálcular el error en el período
error_periodo = np.abs((np.array(periodos_erroneos) - np.array(periodos)) / np.array(periodos))

# Cálcular la media del error
media_error = np.mean(error_periodo)

# Simulación usando la media del error
periodos_con_media_error = [p * (1 + media_error) for p in periodos]

# Visualizar resultados
plt.figure(figsize=(10, 6))
plt.plot(longitudes, periodos, label='Sin error')
plt.plot(longitudes, periodos_erroneos, label='Con error', linestyle=':')
plt.plot(longitudes, periodos_con_media_error, label='Con media del error', linestyle='--')
plt.xlabel('Longitud del péndulo (m)')
plt.ylabel('Período (s)')
plt.title('Efecto del Error en el Período del Péndulo')
plt.legend()
plt.grid(True)
plt.show()
```
