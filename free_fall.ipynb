```python
import matplotlib.pyplot as plt

def simulate_free_fall(time_interval, total_time, gravity=9.81):
    """
    Simulates the free fall of an object, calculating its velocity and fallen
    distance at regular time intervals up to a total time.

    :param time_interval: Time interval for each calculation (in seconds)
    :param total_time: Total time of the simulation (in seconds)
    :param gravity: Acceleration due to gravity (in m/s^2)
    :return: List of tuples (time, velocity, distance)
    """
    results = []
    for t in range(0, total_time + 1, time_interval):
        velocity = gravity * t
        distance = 0.5 * gravity * t**2
        results.append((t, velocity, distance))
    return results

# Example of use
fall_simulation = simulate_free_fall(1, 10)

# Unpack the results
times, velocities, distances = zip(*fall_simulation)

# Create a plot of velocity and distance over time
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(times, velocities, 'b')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity over Time')

plt.subplot(1, 2, 2)
plt.plot(times, distances, 'r')
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.title('Distance over Time')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

def simulate_free_fall_with_air_resistance(time_interval, total_time, mass, area, drag_coefficient, gravity=9.81, air_density=1.225):
    """
    Simulates the free fall of an object with air resistance, calculating its
    velocity and fallen distance at regular time intervals up to a total time.

    :param time_interval: Time interval for each calculation (in seconds)
    :param total_time: Total time of the simulation (in seconds)
    :param mass: Mass of the object (in kg)
    :param area: Cross-sectional area of the object (in m^2)
    :param drag_coefficient: Drag coefficient of the object
    :param gravity: Acceleration due to gravity (in m/s^2)
    :param air_density: Air density (in kg/m^3)
    :return: List of tuples (time, velocity, distance)
    """
    results = []
    velocity = 0
    distance = 0
    for t in range(0, total_time + 1, time_interval):
        air_resistance = 0.5 * drag_coefficient * air_density * area * velocity**2
        net_acceleration = gravity - (air_resistance / mass)
        velocity += net_acceleration * time_interval
        distance += velocity * time_interval
        results.append((t, velocity, distance))
    return results

# Example of use
fall_simulation_with_air = simulate_free_fall_with_air_resistance(1, 20, 1, 0.01, 1.0)

# Unpack the results
times, velocities, distances = zip(*fall_simulation_with_air)

# Create a plot of velocity and distance over time
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(times, velocities, 'b')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity over Time')

plt.subplot(1, 2, 2)
plt.plot(times, distances, 'r')
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.title('Distance over Time')

plt.tight_layout()
plt.show()
```
